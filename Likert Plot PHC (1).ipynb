```python
import pandas as pd

# Load the dataset (adjust filename and path as needed)
df = pd.read_csv("/PHC.csv")

# Preview the dataset
df.head()
```

```python
from pandas.api.types import CategoricalDtype

# Define Likert scale
likert_scale = ["Not Confident", "Slightly Confident", "Neutral", "Confident", "Very Confident"]
cat_type = CategoricalDtype(categories=likert_scale, ordered=True)

# Convert each column to categorical with ordering
for col in df.columns:
    df[col] = df[col].astype(cat_type)
```

```python
# Count frequencies for each category per question
likert_summary = df.apply(lambda x: x.value_counts(normalize=True) * 100).T.fillna(0)

# Ensure correct order of columns
likert_summary = likert_summary[likert_scale]
likert_summary.head()
```

```python
import plotly.graph_objects as go

# Create a stacked bar chart
fig = go.Figure()

# Add each response level as a separate bar
for response in likert_scale:
    fig.add_trace(go.Bar(
        y=likert_summary.index,
        x=likert_summary[response],
        name=response,
        orientation='h'
    ))

fig.update_layout(
    title='Likert Plot of Capacity Level',
    barmode='stack',
    xaxis_title='Percentage (%)',
    yaxis_title='Training Domains',
    template='plotly_white',
    height=600
)

fig.show()
```

```python
import plotly.graph_objects as go

# Create a stacked bar chart
fig = go.Figure()

# Add each response level as a separate bar
for response in likert_scale:
    fig.add_trace(go.Bar(
        y=likert_summary.index,
        x=likert_summary[response],
        name=response,
        orientation='h',
        text=likert_summary[response].round(1).astype(str) + '%',  # Display percentage
        textposition='inside',  # Can be 'outside', 'inside', or 'auto'
        insidetextanchor='start'  # Align text inside the bar
    ))

fig.update_layout(
    title='Likert Plot of Capacity Level',
    barmode='stack',
    xaxis_title='Percentage (%)',
    yaxis_title='Training Domains',
    template='plotly_white',
    height=600
)

fig.show()
```
